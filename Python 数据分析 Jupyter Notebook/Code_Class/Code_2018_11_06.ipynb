# 产生一段时间范围内的时间序列索引 DatetimeIndex 方法

```python
date_list_new = pd.date_range("2018-01-01",periods = 100)
date_list_new
```

```python
date_list_new = pd.date_range("2018-01-01",periods = 100,freq = "W") # 2018-01-03 周日 默认从周日开始到周六结束 一周
date_list_new
```

```python
date_list_new = pd.date_range("2018-01-01",periods = 100,freq = "W-MON") # 设置从周一开始到周日结束
date_list_new
```

```python
date_list_new = pd.date_range("2018-01-01",periods = 100,freq = "H")
date_list_new
```

```python
date_list_new = pd.date_range("2018-01-01",periods = 100,freq = "5H")
date_list_new
```

```python
s2 = Series(np.random.rand(100), index = date_list_new)
s2
```

# 时间序列数据的采样和画图

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
t_range = pd.date_range("2018-01-01","2018-12-31")
t_range
```

```python
s1 = Series(np.random.randn(len(t_range)),index = t_range)
s1
```

```python
s1["2018-01"].mean() # 计算2018年1月的平均值
```

```python
s1_month = s1.resample("M").mean() # 按月进行采样 采样也成取样，指把时间域或空间域的连续量转化成离散量的过程，取平均值
s1_month
```

```python
s1_month.index # index 为每月的最后一天
```

```python
s1.resample("H").ffill() # forward fill 也可以进行插值，按小时插值
```

```python
s1.resample("H").bfill() # backworad fill
```

```python
t_range = pd.date_range("2018-01-01","2018-12-31",freq = "H")
t_range
```

```python
stock_df = DataFrame(index = t_range) # 创建DataFrame的时间索引
print(type(stock_df.index))
stock_df.head()
```

```python
stock_df["阿里巴巴"] = np.random.randint(80,160,size = len(t_range))
stock_df["腾讯"] = np.random.randint(30,50,size = len(t_range))
stock_df.head()
```

```python
stock_df.plot()
```

```python
import matplotlib.pyplot as plt

plt.show()
```

```python
weekly_df = DataFrame()
weekly_df["阿里巴巴"] = stock_df["阿里巴巴"].resample("W").mean()
weekly_df["腾讯"] = stock_df["腾讯"].resample("W").mean()
weekly_df.head()
```

```python
weekly_df.plot()
plt.show()
```
