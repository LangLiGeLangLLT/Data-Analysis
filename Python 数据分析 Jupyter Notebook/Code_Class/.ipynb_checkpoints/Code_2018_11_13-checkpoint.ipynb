# 透视表功能实现

```python
df.sort_values("arr_delay",ascending = False)[:10]
```

```python
df["cancelled"].value_counts()
```

```python
df["delayed"] = df["arr_delay"].apply(lambda x:x>0)
df.head()
```

```python
delay_data = df["delayed"].value_counts()
delay_data
```

```python
delay_data[1]/(delay_data[0]+delay_data[1])
```

```python
delay_group = df.groupby(["unique_carrier","delayed"])
delay_group
```

```python
delay_group.groups
```

```python
delay_group.size()
```

```python
type(delay_group.size())
```

```python
df_delay = delay_group.size().unstack()
df_delay
```

```python
import matplotlib.pyplot as plt

df_delay.plot()
```

```python
df_delay.plot(kind = "barh",figsize = [16,6])
```

```python
df_delay.plot(kind = "barh",figsize = [16,6],stacked = True,colormap = "winter")
```
